```python
# Age vs Fare scatterplot
sns.scatterplot(x='age', y='fare', data=df, hue='survived', alpha=0.5, palette='Set2')
plt.title('Age vs Fare (coloured by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare (£)')
plt.show()
```

```python

# Fare by passenger class
sns.boxplot(x='pclass', y='fare', data=df, palette='Set2')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare (£)')
plt.show()
```

```python
corr=df[['survived', 'age', 'fare', 'pclass', 'sibsp', 'parch']].corr()
sns.heatmap(corr,annot=True,fmt='.2f',cmap='coolwarm')
plt.show()
```

## Key Findings

**Finding 1:** Female passengers had a significantly higher survival rate (~74%)
compared to male passengers (~19%). Gender was the strongest survival predictor.

**Finding 2:** First class passengers had the highest survival rate (~63%),
followed by second class (~47%), and third class (~24%).
Higher fare = better access to lifeboats.

**Finding 3:** Fare is right-skewed with significant outliers.
A small number of first-class passengers paid extremely high fares.

**Finding 4:** Age and fare show no strong linear relationship (r ≈ 0.09).
Fare is driven more by passenger class than by age.

**Finding 5:** Pclass and fare are strongly negatively correlated (r ≈ -0.55).
As class number increases (1→3), fare decreases — as expected.
